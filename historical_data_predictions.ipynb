```python
# wrote a function that takes the training set, testing set, a model, a placeholder for predictions. returns a dataframe of the actual values
# and the target values.

def predict(train, test, predictions, model):
    model.fit(train[predictors], train["target"])
    predictions = model.predict(test[predictors])
    predictions = pd.Series(predictions, index=test.index, name="predictions")
    combined_values = pd.concat([test["target"], predictions], axis=1)
    return combined_values
```

```python
# furthur simulating the model multiple times on growing slices of data, to show how the model would perform in the 
# real world.

def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
        
    return pd.concat(all_predictions)
```

```python
predictions = backtest(stock_data, model, predictors)
```

```python
# counts the amount of times the model predicted if it would go up or down

predictions["predictions"].value_counts()
```

```python
# the model correctly predicted the movement of future stock price about 55.17% of the time. 

precision_score(predictions["target"], predictions["predictions"])
```

```python
# showing the percentage of predicted stocks that went up or down.

predictions["target"].value_counts() / predictions.shape[0]
```

```python
# defined time horizons to analyze various trends. By looping through the horizons index, a moving average, mean reversion, 
# and trend (the number of times the stock went up in the last horizon days) is calculated for each horizon; which is then added to 
# new_predictors.

horizons = [2,5,60,250,1000]

new_predictors = []

for horizon in horizons:
    rolling_avg = stock_data.rolling(horizon).mean()

    ratio_col = f"close_ratio_{horizon}"
    stock_data[ratio_col] = stock_data['4. close'] / rolling_avg['4. close']

    trend_col = f"trend_{horizon}"
    stock_data[trend_col] = stock_data.shift(1).rolling(horizon).sum()["target"]

    new_predictors += [ratio_col, trend_col]
```

```python
# initializing another model, where I found n_estimators = 320 and min_samples_split = 50 to output the best predictive
# accuracy. 

model = RandomForestClassifier(n_estimators=320, min_samples_split=50, random_state=1)
```

```python
# modifying the prediction function earlier, with the difference of defining of the probability that the prediction is 
# correct to be above 0.6 to ensure accuracy. 

def predict(train, test, predictions, model):
    model.fit(train[predictors], train["target"])
    predictions = model.predict_proba(test[predictors])[:,1]
    predictions[predictions >= .6] = 1
    predictions[predictions < .6] = 0
    predictions = pd.Series(predictions, index=test.index, name="predictions")
    combined_values = pd.concat([test["target"], predictions], axis=1)
    return combined_values
```
